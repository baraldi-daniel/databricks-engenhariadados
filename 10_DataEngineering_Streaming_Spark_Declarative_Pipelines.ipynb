# Data Engineering with Spark Declarative Pipelines

Spark Declarative Pipelines is a way of working that simplifies much of the steps executed in Notebooks. The idea behind this notebook is to use SDP as a way to increase productivity and simplify steps.

Create a new catalog called pipelines_sdp and a schema called schema_sdp

```python
# Create a new catalog and a new schema
spark.sql("CREATE CATALOG pipelines_sdp MANAGED LOCATION 'abfss://container@baraldistorage.dfs.core.windows.net/pipelines_sdp'")
spark.sql("USE CATALOG pipelines_sdp")
spark.sql("CREATE SCHEMA IF NOT EXISTS schema_sdp")
```
