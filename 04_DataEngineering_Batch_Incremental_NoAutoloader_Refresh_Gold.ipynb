## Load Data for Gold

```python
from pyspark.sql import functions as F
```

Customer Dimension

```python
# Reading data
df_silver=spark.read.table("medallion.silver.dim_customer")

# Treatment for gold
df_silver=df_silver.filter(df_silver.is_current == True)

df_silver \
  .write \
  .mode("overwrite") \
  .saveAsTable("medallion.gold.dim_customer")

# Creating view
spark.sql("""
  CREATE OR REPLACE VIEW medallion.gold.vw_dim_customer AS
  SELECT *
  FROM medallion.gold.dim_customer
""")
```
